# Ejercicio 3. Grupo en el zoo

## Enunciado

Un zoo fija el precio de la entrada según la edad del visitante, así:
- los niños de 2 años o menos no pagan
- los niños de 3 a 12 pagan 14€
- A partir de 13 años se considera entrada de adulto por un valor de 23 €
- Los jubilados, 65 o más años pagan 18 €

Debes hacer un programa que pida las edades del grupo que va a visitar el zoo y devolver el precio del grupo y el detalle por edades.

El programa devolverá el cálculo y el detalle cuando se no se introduzcan más edades, es decir se introduzca ''.

Debes asegurarte de que la edad sea un número entero y positivo.



## Ayuda a la implementación

Puedes ayudarte de estructuras de datos (listas o tuplas) para almacenar las entradas y edades mientras se informan.


```python
""" 
pida las edades del grupo que va a visitar el zoo y devolver el precio del grupo y el detalle por edades.
El programa devolverá el cálculo y el detalle cuando se no se introduzcan más edades, es decir se introduzca ''.

.- <=2  0€
.- 3-12 14€
.- 13-64 23€
.- >= 65 18€
"""
#Constants 
from enum import Enum
class Age_Group (Enum):
    BABY = 0
    CHILD = 1
    ADULT = 2
    JUBILATED = 3

class price_tickets(Enum):
    BABY = 0
    CHILD= 14
    ADULT = 23
    JUBILATED = 18

# Functions 

def is_valid_age(user_string:str)->bool:
    """ 
    Recibe la entrada de usuario y determina si es un valor válido para ser tratado
    """
    result = False
    try: 
        if int(user_string) > 0:
            result = True
    except ValueError:
        print('No valid data...')
    return result
        
def update_key(age:int, ticket_list:dict[int])->dict[int]:
    if age in ticket_list:
        ticket_list[age] +=1
        
    else: 
        ticket_list[age] = 1
    ticket_list.update(ticket_list)
    return ticket_list

total_price =0
tickets_group = {}
while True: 
    age_for_ticket = input('Please, insert age of ther person that you want buy a ticket ')
    if is_valid_age(age_for_ticket):
        if int(age_for_ticket) < 3:
            tickets_group = update_key(Age_Group.BABY, tickets_group)
            total_price+= price_tickets.BABY.value
        elif int(age_for_ticket) <13 :
            tickets_group = update_key(Age_Group.CHILD, tickets_group)
            total_price+=price_tickets.CHILD.value
        elif int(age_for_ticket) < 65:
            tickets_group = update_key(Age_Group.ADULT, tickets_group)
            total_price+=price_tickets.ADULT.value
        else:
            tickets_group = update_key(Age_Group.JUBILATED, tickets_group)
            total_price+=price_tickets.JUBILATED.value
    if age_for_ticket == '.':
        break
for key, value in tickets_group.items(): 
    print(f'{value} tickets for {key}')
print(f'Total cost: {total_price}€' )
```
