# Ejercicio 4. AENIGMA programable

## Enunciado

Partiendo de la máquina aenigma, vemos que está limitada porque siempre funciona con la misma clave. Conviertela en una versión mejor. Anádele la posibilidad de indicar al encender una nueva clave. Y en lo que era apagar (escribir '****') en lugar de apagarla, que pida una nueva clave.

Ahora si quieres apagar la máquina, deberás introducir la clave 0

Debes controlar que la clave sea un número entero mayor o igual que cero, si no lo fuera avisar al usuario con un mensaje de error.


## Ayuda a la implementación

Básicamente debes encontrar la forma de implementar sobre el bucle de funcionamiento básico de aenigma, otro bucle que controle la clave y el apagado.
