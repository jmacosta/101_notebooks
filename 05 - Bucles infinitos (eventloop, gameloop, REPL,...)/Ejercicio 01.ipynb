# Ejercicio 1. Acceso por pin

## Enunciado

Debes crear un programa que pida un pin de acceso de 4 cifras y solo permita el acceso cuando se adivine el pin.




## Ayuda a la implementación

- Recuerda que input siempre devuelve strings aunque se hayan escrito números.


```python
""" 
Programa que pide un pin y solo se permite el accesos cuando se adivine
"""
ACCESS_PIN = 4252
while True: 
    try: 
        user_pin = int(input('Please insert pin '))
    except ValueError:
        pass
    if ACCESS_PIN != user_pin: 
        print('Access denied, Incorrect pin...')
    else: 
        print ('Access granted')
        break
    
```
