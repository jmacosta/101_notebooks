# Ejercicio 6. Mini calculadora de números

## Enunciado

Debes hacer un programa que presente las siguientes opciones:
```
S. Sumar
R. Restar
M. Multiplicar
D. Dividir
X. Salir
Introduce opción: |
```

El programa debe esperar una opción. Responde tanto a minúsculas como mayúsculas.

Si se introduce una opción distinta de las mostradas debe mostrar el mensaje "OPCION INCORRECTA" y debe volver a imprimir las opciones y esperar entrada

Para el resto de opciones:
- Si introduce S o s:
    - Debe pedir numeros hasta que no se introduzca nada ("").
    - Cuando se de al enter sin nada. Debe mostrar el resultado de la suma y volver al menú, imprimiéndolo de nuevo y volviendo a esperar opción
    - Si al pedir los números se introduce algun valor no numérico debe presentarse el mensaje "NUMERO NO VALIDO" y seguir con el ciclo

- Si se introduce R o r: 
    - Debe pedir dos números y controlar que son válidos
    - Debe mostrar la resta de ambos y mostrar el mensaje "PULSE UNA TECLA" para volver al menú

- La opción M o m funciona como la suma pero con multiplicaciones

- La opción D o d funciona como la resta pero con divisiones.
