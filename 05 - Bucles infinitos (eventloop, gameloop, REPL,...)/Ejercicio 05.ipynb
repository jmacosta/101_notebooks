# Ejercicio 5. La taquilla del zoo

## Enunciado

Partiendo del ejercicio de las entradas del zoo, crear un programa que tras calcular el precio de las entradas de un grupo pregunte si hay un nuevo grupo. De ser así (S) se reinicia el grupo y se vuelven a pedir edades y calcular precio de las entradas. Si se selecciona que no hay nuevo grupo se acaba el programa.

## Ayuda a la implementación

Básicamente se trata de encontrar como anidar el bucle que ya tienes en el bucle de pedir nuevo grupo.
