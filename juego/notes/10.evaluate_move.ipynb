## ¿Y si una de las dos opciones resultan ser INVALID o QUIT?

Realmente no debería de ocurrir jamás y tal como está el programa, no ocurrirá. Sin embargo podría ocurrir el el futuro por accidente al cambiar algo en las funciones que se ejecutan antes que la de evaluación.

Eso sería un **pánico** que deja el sistema en un estado desconocido. Nuestra función no tiene como comparar una jugada con una de esas opciones.

Haremos un `assert` al principio de la función para validar ambos parámetros. Si pasamos el `assert` es que los parámetros son correctos y el resto del código se puede ejecutar.

El `assert` en caso de fallar, lanzará una excepción de tipo `AssertionError`. Ésta, al no ser capturada en `evaluate_move`, subirá a la función que la ha llamado (`game_loop`). Como ahí tampoco es capturada, seguirá subiendo y será capturada en el `try` que llama a `game_loop`(los imsebelen).



```python
from enum import Enum

class UserChoice(Enum):
    INVALID = -1
    PAPER = 0
    ROCK = 1
    SCISSORS = 2
    QUIT = 3

def evaluate_move(user_choice, computer_choice):
    """
    Recibe dos jugadas, determina cual ha ganado y devuelve un mensaje con el resultado.
    Por ejemplo: recibe Papel y Piedra, y devuelve "Papel envuelve Piedra"
    """
    assert user_choice != UserChoice.INVALID and user_choice != UserChoice.QUIT
    assert computer_choice != UserChoice.INVALID and computer_choice != UserChoice.QUIT

    result = ''
    if user_choice == UserChoice.PAPER:
        if computer_choice == UserChoice.PAPER:
            result = "It's a tie!"
        elif computer_choice == UserChoice.ROCK:
            result = "You win! Paper covers Rock."
        elif computer_choice == UserChoice.SCISSORS:
            result = "You lose! Scissors cut Paper"
        
    elif user_choice == UserChoice.ROCK:
        if computer_choice == UserChoice.ROCK:
            result = "It's a tie!"
        elif computer_choice == UserChoice.PAPER:
            result = "You lose! Paper covers Rock."
        elif computer_choice == UserChoice.SCISSORS:
            result = "You win! Rock smashes Rock"

    elif user_choice == UserChoice.SCISSORS:
        if computer_choice == UserChoice.SCISSORS:
            result = "It's a tie!"
        elif computer_choice == UserChoice.ROCK:
            result = "You lose! Rock smashes Scissors."
        elif computer_choice == UserChoice.PAPER:
            result = "You win! Scissors cut Paper"
    
    return result
```
