# Ejercicio 7. Aproximación a pi

## Enunciado

Una forma de aproximarse al valor de pi es sumar cuantos más términos mejor de la siguiente secuencia:

$\pi = 3 +\frac{4}{2 \cdot 3 \cdot 4}-\frac{4}{4 \cdot 5 \cdot 6}+\frac{4}{6 \cdot 7 \cdot 8}-\frac{4}{8 \cdot 9 \cdot 10} + ...$

Construye un programa que presente las primeras 15 aproximaciones del número, para que se vea como se va acercando. Consideraremos la primera aproximación la primera suma.

$3 +\frac{4}{2 \cdot 3 \cdot 4}$

Debes mostrar el resultado de esta y de 14 iteraciones más, cada una en su línea.

## Ayuda a la implementación

Intenta ayudarte de range para hacerlo, recuerda que range puede incluir saltos.

```python
""" 
Construye un programa que presente las primeras 15 aproximaciones de Pi
"""

def approach_pi(index):
    """ 
    Calculo el valor de la aproximación a pi segun el numero recibido
    """
    return 4/(index*(index+1)*(index+2))

pi = 3 + approach_pi(2)
print(pi)
signal = False
for index in range(4,16*2,2):
    if signal == True:    
        pi += approach_pi(index)
        signal = False
    else:
        pi -= approach_pi(index)
        signal = True
    print(pi)
    

```
