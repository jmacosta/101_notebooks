# Ejercicio 3. Sin repeticiones

## Enunciado

Necesitamos una función que recibirá una lista con 6 elementos (cadenas o números) y devolverá la misma lista sin repeticiones. Con los elementos en el orden original



## Ayuda a la implementación

Recuerda que puedes saber si un elemento está en una lista con el operador `in`

```python
def without_repeat(list_things):
    """ 
    recibe una lista con 6 elementos y devolvera la misma lista sin repeticiones 
    """
    result = []
    for element in list_things:
        if element not in result:
            result.append(element)
    return result        
```

```python
without_repeat([2,3,'45',2,4,6])
```
