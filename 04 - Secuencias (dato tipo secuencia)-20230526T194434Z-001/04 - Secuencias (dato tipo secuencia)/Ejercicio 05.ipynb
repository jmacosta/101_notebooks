# Ejercicio 5. Extraer dígitos de un número cualquiera

## Enunciado

Basándote en la solución del ejercicio 7 del bloque de booleanos, que juega con cociente y resto de la división entre potencias de 10 para sacar los tres dígitos de un número de tres cifras. Construye una función que te devuelva una lista ordenada con los dígitos de un número entero ignorando su signo talque:
- `digitos(2837) -> [2, 8, 3, 7]`
`

## Ayuda a la implementación

Recuerda que si quieres extraer la cifra que equivale a millares debes obtener la división entera `numero // 1000`. Su resultado será la cifra y debes, luego, procesar el resto de dicha división `numero % 1000`. Con el resto debes proceder igual pero ahora con 100, luego con 10 y el resto de esta última división será el número de unidades.

```python
def digits(number):
    """ 
    devuelva una lista ordenada con los dígitos de un número entero ignorando su signo
    """
    result = []
    work_number = number
    for index in range(len(str(number))):
        result.append(work_number //  10** (len(str(number))-index-1))
        work_number = work_number %  10** (len(str(number))-index-1)
    return result
```
