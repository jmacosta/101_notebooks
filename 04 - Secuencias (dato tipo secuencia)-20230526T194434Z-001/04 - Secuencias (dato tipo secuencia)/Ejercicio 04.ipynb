# Ejercicio 4. Por debajo, en y por encima de la media

## Enunciado

Crea una función que de una lista de 6 números enteros, separe en tres listas los números que están por debajo, por encima y son iguales a la media (si hay alguno) y que mantenga el orden de entrada


## Ayuda a la implementación

Mira el ejercicio 2

```python
def under_middle_over(list_numbers):
    """ 
    Recibe una lista y los separa en menor igual o superior
    """
    acum = 0
    for element in list_numbers:
        acum += element
    print (f'{"Over":<10} {"Middle":^10}{"Under":>10}')
    
    for element in list_numbers:
        if element > acum /len(list_numbers):
            print (f'{element:<10}{"":^10}{"":>10}')
        elif element < acum /len(list_numbers):    
            print (f'{"":<10}{"":^10}{element:>10}')
        else: 
            print (f'{"":<10}{element:^10}{"":>10}')
```

```python
under_middle_over([1,3,5,6,0])
```
