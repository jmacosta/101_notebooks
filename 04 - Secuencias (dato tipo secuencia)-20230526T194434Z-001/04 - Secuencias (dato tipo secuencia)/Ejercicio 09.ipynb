# Ejercicio 9

## Enunciado. Vasos reciclados

Una fábrica produce vasos reciclados partiendo de vasos usados. Conociendo el número de vasos reciclados y cuantos de ellos se necesitan para obtener uno reciclado, crea una función que te diga el máximo de vasos que puedes conseguir de una cantidad cualquiera de vasos reciclados, reciclándolos sucesivas veces.



## Ayuda a la implementación

Imagina que se parte de 70 vasos reciclados y que se necesitan 4 vasos usados para crear uno reciclado. Nuestra función debería realizar lo siguiente

| #n   |  recogidos   | fabricados | sobrantes |
|:------:|--------------|------------|-----------|
| 1 | $$70$$ | $$\frac{70}{4} = 17$$ | $$2$$ |
| 2 | $$17+2$$ | $$\frac{19}{4} = 4$$ | $$3$$ |
| 3 | $$4+3$$ | $$\frac{7}{4} = 1$$ | $$3$$ |
| 4 | $$1+3$$ | $$\frac{4}{4}= 1$$ | $$0$$ |
| 5 | $$1$$ | no hay bastantes vasos para reciclar | $$0$$ |

El total de vasos reciclados es 17 + 4 + 1 + 1 = 23

Dado que no se puede saber de antemano el número de repeticiones es un buen ejercicio para usar while


```python
def recycling_glasses(glases):
    """
    devuelve el máximo de vasos que puedes conseguir de una cantidad cualquiera de vasos reciclados, reciclándolos sucesivas veces.
    """
    recicled_glases = 0
    for_recycle = glases
    temp = 0
    while for_recycle >= 4:
        temp = for_recycle //4 
        recicled_glases += temp
        for_recycle = temp + for_recycle % 4
        
    return recicled_glases
recycling_glasses(70)    
```
