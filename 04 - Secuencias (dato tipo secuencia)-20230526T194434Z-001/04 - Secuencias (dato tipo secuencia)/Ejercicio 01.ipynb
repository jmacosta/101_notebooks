# Ejercicio 1. Elementos de una sucesión menores o iguales a 100

## Enunciado

Crea las funciones que devuelvan las listas con los elementos de las siguientes sucesiones que sean menores o iguales a 100. Utiliza para ello sólo range.

- Primera sucesión: 1, 4, 7, 10, 13,...
- Segunda sucesión: -12, -7, -2, 3, 8,...



## Ayuda a la implementación

Repasa los tres parámetros que definen un range

```python
def sequence_by_3():
    """_summary_ Secuencia de numeros hasta 100 de 3 en 3    
    """
    result =[]
    for element in range(1,100,3):
        result.append(element)
    return result
```

```python
sequence_by_3()
```

```python
def sequence_by_5():
    """_summary_ Secuencia de numeros hasta 100 de 5 en 5    
    """
    result =[]
    for element in range(-12,100,5):
        result.append(element)
    return result
```

```python
sequence_by_5()
```
