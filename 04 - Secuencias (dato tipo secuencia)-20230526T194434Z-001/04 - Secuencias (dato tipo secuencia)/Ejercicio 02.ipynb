# Ejercicio 2. Boletín de notas

## Enunciado. 

Crea una lista con 5 asignaturas. Después crea un programa que te pida la nota de cada asignatura mostrando el mensaje "Nota de <Asignatura>?"

El programa debe, una vez introducidas todas la notas mostrar el boletín de notas según el siguiente esquema:
```

--------------------------------------------
Nombre de asignatura 1:                 10.0
El nombre de la segunda:                 4.5
nombre corto de 3                        6.8
la cuarta                                7.5
ultima asignatura                        5.0
--------------------------------------------
NOTA MEDIA                               6.8

```
Es decir debes mostrar los nombres y notas ajustados, teniendo en cuenta que los nombres de las asignaturas tendrán longitudes diferentes. E igual con las notas.

Debes además calcular y mostrar la media de las 5 notas y ajustarlo a un único decimal

## Ayuda a la implementación

Puedes utilizar las f-strings de python para dar formato a las cadenas. Puedes ver la información de uso [aquí](https://docs.python.org/3/tutorial/inputoutput.html#tut-f-strings) y detalles sobre el minilenguage de formateo [aquí](https://docs.python.org/3/library/string.html#formatspec)

````python
"""_summary_
    Crea una lista con 5 asignaturas. Después crea un programa que te 
    pida la nota de cada asignatura 
"""
def request_grade(subject):
    """
    Recibe una asignatura y solicita la nota 
    """
    grade = None
    while True:
        grade = input('Nota de '+ subject + ' ? ')
        if float(grade) > 0 and float(grade) < 10:
            break
    return grade

def calculate_gpa(grades):
    """ 
    Recibe una lista de notas y calcula la nota media
    """
    result = 0
    for element in grades: 
        result += element
    return result / len(grades)    

def present_report_card (subjects, grades, gpa):
    """ 
    Recibe las asignaturas, notas y nota media y los presenta
    """
    index =0
    print ('{:-^50}'.format(''))
    for element in subjects:
        print (f'{element:46}{grades[index]:.2f}')
        index += 1
    print ('{:-^50}'.format(''))
    print(f'{"Nota media":45} {gpa:.2f}')
    return None
    





subjects= ['Mathematics', 'Science', 'English', 'History', 'Physical Education']
grades=[]
gpa = 0
for element in subjects: 
    grades.append(float(request_grade(element)))
gpa = calculate_gpa(grades)  
present_report_card(subjects, grades, gpa)  
        

````
